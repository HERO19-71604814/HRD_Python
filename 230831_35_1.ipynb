# 텍스트 파일 읽기

```python
# 파일변수명 = open('파일명','r')
# 변수명 = 파일변수명.read()
# replace(): 다른 텍스트로 대처

file = open('data/문재인대통령.txt','r',encoding='utf-8')
contents = file.read()
contents = contents.replace('\n', ' ')
contents = contents.replace('!', '')
contents = contents.replace('.', '')
words = contents.split(' ')

word_dict = {}

for word in words:
    if word not in word_dict:
        word_dict[word] = 0
    word_dict[word] += 1 
print(word_dict)
```
