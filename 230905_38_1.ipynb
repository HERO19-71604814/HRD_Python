# 파일처리(326p)

```python
import csv
file = open("data/승하차인원.csv")
data = csv.reader(file)
header = next(data)
print(header)
st_conut = []

for row in data:
    row[2:] = map(int, row[2:])
    print(row)
    st_conut.append(row)
```

## 역별 승하차 인원 합계

```python
st_raid_quit = {}

for row in st_conut:
    if row[0] not in st_raid_quit: # 승차
        st_raid_quit[row[0]] = {}
    if row[1] not in st_raid_quit[row[0]]: # 하차
        st_raid_quit[row[0]][row[1]] = 0
    st_raid_quit[row[0]][row[1]] += sum(row[2:])

st_raid = {} # 승차
st_quit = {} # 하차
for key, value in st_raid_quit.items():
#     print(key, ':', value)
    st_raid[key] = value['승차']
    st_quit[key] = value['하차']
# print('- 승차')
# print(st_raid)
# print('- 하차')
# print(st_quit)

st_raid_sort = sorted(st_raid.items(), key = lambda item : item[1], reverse=True)
print('-- 대전 도시 철도 승차 인원 --')
for key in st_raid_sort:
    print(key)

print('-- 대전 도시 철도 하차 인원 --')
st_quit_sort = sorted(st_quit.items(), key = lambda item : item[1], reverse=True)
for key in st_quit_sort:
    print(key)
    
```
