```python
with open('data/info.txt', 'r') as file:
    data = file.readlines()
    for info in data:
        info = info.replace('\n', '')
        info = info.split(', ')
        bmi = int(info[2]) / (int(info[1]) / 100) **2
        result = ""
        if bmi >= 25:
            result = "비만"
        elif bmi < 18.5:
            result = "미달"
        else:
            result = "정상"
        print(info, result)
            
```

```python
with open('data/info.txt', 'r') as file:
    data = file.readlines()
    info_list = []
    for i in range(len(data)):
        data[i] = data[i].replace('\n', '')
        split_info = data[i].split(', ')
        split_info[1:] = map(int, split_info[1:])
        bmi = split_info[2] / ((split_info[1] / 100) ** 2) 
        result = ""
        if bmi >= 25:
            result = "비만"
        elif bmi < 18.5:
            result = "미달"
        else:
            result = "정상"
        split_info.append(result)
        info_list.append(split_info)
```

```python
bmi_count = {}
for info in info_list:
    if info[3] not in bmi_count:
        bmi_count[info[3]] = 0
    bmi_count[info[3]] += 1
print(bmi_count)
```

### 키가 150대, 160대,,,, 190대별 인원 수 구하기

```python
#  height_list = ['150대', '160대', '170대', '180대', '190대']
dict_height = {}

for info in info_list:
    h = int(info[1] / 10)
    key = str(h) + '0대'
   
    if key not in dict_height:
        dict_height[key] = 0
    dict_height[key] += 1
    
for key, value in dict_height.items():
    print(key, value)
```
