# ※ 바이오 리듬

```python
# import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

from math import sin, radians
from datetime import datetime

today = datetime.now()
birth = datetime(1999, 2, 16)
print("현재일 : ", today)
print("생년월일 : ", birth)

days = (today - birth).days
print("살아온 날: ", days)

# 각 바이오리듬 주기
ph = 23 # 신체
em = 28 # 감성
im = 33 # 지성

# 각 주기별 각도
ph_r = (360 / ph) 
em_r = (360 / em)
im_r = (360 / im) 

# 살아온 날의 주기별 나머지 값
ph_d = days % ph
em_d = days % em
im_d = days % im

# 저장 공간
ph_sin = []
em_sin = []
im_sin = []

# 30일간 바이오리듬값 구하기
for i in range(0, 61):
    ph_sin.append(sin(radians((ph_d + i) * ph_r)))
    em_sin.append(sin(radians((em_d + i) * em_r)))
    im_sin.append(sin(radians((im_d + i) * im_r)))
    
plt.figure(figsize=(15, 8))
plt.title('Biorhythm', fontsize=20)
plt.plot(ph_sin, color='blue', label='Physical')
plt.plot(em_sin, color='red', label='emotion')
plt.plot(im_sin, color='green', label='intelligent')
plt.legend(loc=0)
plt.show()
```

```python
import random as ran
print(ran.random()) # 0 ~ 0.999999사이의 수
print()
print(ran.randint(1, 5)) # 1 부터 5까지의 랜덤 정수 
print()
print(ran.uniform(10, 50)) # 10 부터 50까지의 랜덤 실수
print()
print(ran.randrange(10)) # 0 부터 10사이의 랜덤 수
print(ran.randrange(1, 10)) # 1 부터 10사이의 랜덤 수
print()
print(ran.choice([1,2,3,4,5,6,7,8,9])) # 리스트에서 랜덤으로 하나 선택
print()
print(ran.choices([1,2,3,4,5,6,7,8,9], k=10)) # 리스트에서 랜덤으로 여러개 선택
print()
print(ran.sample([1,2,3,4,5,6,7,8,9], 3)) # 리스트에서 랜덤으로 여러개 선택
print()
# print(random.sample(set([1,2,3,4,5,4, 5, 6, 7,6,7,8,9]), 5)) Error
```
