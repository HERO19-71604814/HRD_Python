```python
from urllib import request 
from bs4 import BeautifulSoup

target = request.urlopen("http://www.naver.com")
output = target.read()

html = BeautifulSoup(output, "html.parser")
print(html)
```

```python
url = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
target = request.urlopen(url)
soup = BeautifulSoup(target, "html.parser")
title = soup.select('title')

# print(soup)
print(title[1].string)
location = soup.select('location')
# print(location)

for loc in location:
    print('=============================================')
    print('지역 : ',loc.select_one('city').string)
    print('=============================================')
    data = loc.select('data')
    for wea in data:
        print('시간 : ', wea.select_one('tmef').string)
        print('날씨 : ', wea.select_one('wf').string)
        print('최저 : ', wea.select_one('tmn').string)
        print('최고 : ', wea.select_one('tmx').string)
        print('강수량 : ', wea.select_one('rnst').string)
        print()
    
```
