# 위키백과의 방탄소년단의 대한 데이터를 파이썬으로 출력하기

```python
from bs4 import BeautifulSoup as bs
from urllib import request 

url = "https://ko.wikipedia.org/wiki/%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8"
target = request.urlopen(url)
soup = bs(target.read(), "html.parser")

h3 = soup.select('h3')

for h in h3:
    title = h.find('span',({'class' : 'mw-headline'})).string
    if '년' in title:
        print(title)
        contents = h.next_sibling.next_sibling
        print(contents)
```
