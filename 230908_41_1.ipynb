# 오피넷에서 휘발유에 대한 데이터 가져오기

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
driver = webdriver.Chrome()
# url = "https://www.opinet.co.kr/user/main/mainView.do"
url = "https://www.opinet.co.kr/searRgSelect.do"
driver.get(url)
# time.sleep(2)
# driver.find_element(By.XPATH, '//*[@id="quick_ul"]/li[2]/a/span').click()
time.sleep(3)
driver.find_element(By.XPATH, '//*[@id="SIDO_NM0"]/option[7]').click()
time.sleep(1)
dict_gu = {}
for i in range(2, 7):
    driver.find_element(By.XPATH, '//*[@id="SIGUNGU_NM0"]/option['+str(i)+']').click()
    gu = driver.find_element(By.XPATH, '//*[@id="SIGUNGU_NM0"]/option['+str(i)+']').text
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="searRgSelect"]/span').click()
    time.sleep(2)
    oil_cnt = driver.find_elements(By.XPATH, '//*[@id="body1"]/tr')
    print('===================')
    print(gu)
    print('===================')
    dict_gu[gu] = {}
#     print(len(oil_cnt))
    for j in range(1, len(oil_cnt)+1):
        comp = driver.find_element(By.XPATH, '//*[@id="body1"]/tr['+str(j)+']/td[1]/a').text
        gas = driver.find_element(By.XPATH, '//*[@id="body1"]/tr['+str(j)+']/td[2]').text
        via = driver.find_element(By.XPATH, '//*[@id="body1"]/tr['+str(j)+']/td[3]').text
        print(comp, gas, via)
        dict_gu[gu][comp] = [gas, via]
        
print(dict_gu)        
```

- 대전 구별 휘발유, 경유값 dict : dict_gu 
- 대전시 전체 휘발유, 경유값 평균
- 휘발유, 경유값 총액과 주유소 개수

```python
oil_sum = 0
oil_cnt = 0
via_sum = 0
via_cnt = 0
for key, value in dict_gu.items():
    for subkey, subval in value.items():
        oil_sum += int(subval[0])
        oil_cnt += 1
        via_sum += int(subval[1])
        via_cnt += 1

print('대전시 휘발유 평균값 : ', oil_sum / oil_cnt)
print('대전시 경유 평균값 : ', via_sum / via_cnt)

```

## 구별 통계

```python
for key, value in dict_gu.items():
    oil_sum = 0
    oil_cnt = 0
    via_sum = 0
    via_cnt = 0
    
    for subkey, subval in value.items():
        oil_sum += int(subval[0])
        oil_cnt += 1
        via_sum += int(subval[1])
        via_cnt += 1

    print('대전시 {} 휘발유 평균값 : {}'.format(key, int(oil_sum / oil_cnt)))
    print('대전시 {} 경유 평균값 : {}'.format(key, int(via_sum / via_cnt)))
    print()

```

## 구별 휘발유값이 가장 싼 주유소와 값              

```python
for key, value in dict_gu.items(): # 지역, 주유소와 가격
    min_value = 9999
    max_value = 0
    
    for subkey, subval in value.items(): # 주유소, 가격
        if int(subval[0]) < min_value:
            min_value = int(subval[0]) # subval[0] = 휘발유 => 휘발유가격이 작은 것 
            min_comp = subkey 
        if int(subval[0]) > max_value:
            max_value = int(subval[0]) # subval[0] = 휘발유 => 휘발유가격이 큰 것
            max_comp = subkey
            
    print('대전시 {} 가장 싼 휘발유 값은{}이고 주유소는 {}이다.'.format(key, min_value, min_comp))
    print('대전시 {} 가장 비싼 휘발유 값은{}이고 주유소는 {}이다.'.format(key, max_value, max_comp))
    print()
```

## 구별 휘발유값이 가장싼 주유소와 값

```python
for key, value in dict_gu.items(): # 지역, 주유소와 가격
    min_value = 9999
    max_value = 0
    min_comp = []
    max_comp = []
    
    for subkey, subval in value.items(): # 주유소, 가격
        if int(subval[0]) < min_value:
            min_value = int(subval[0]) # subval[0] = 휘발유 => 휘발유가격이 작은 것 
        if int(subval[0]) > max_value:
            max_value = int(subval[0]) # subval[0] = 휘발유 => 휘발유가격이 큰 것
            
    for subkey, subval in value.items(): # 주유소, 가격
        if min_value == int(subval[0]):
            min_comp.append(subkey)
        if max_value == int(subval[0]):
            max_comp.append(subkey)        
    print('{}의 가장 싼 휘발유 값은 {}이고 주유소는 {}이다.'.format(key, min_value, min_comp))
    print('{}의 가장 비싼 휘발유 값은 {}이고 주유소는 {}이다.'.format(key, max_value, max_comp))
    print()
```
