# n_gram

```python
from collections import Counter

file = open('data_speech/문재인2021.txt', 'r')
conts = file.read()
# print(conts)
conts = conts.replace('\n', ' ')
conts = conts.replace('.', '')
conts = conts.replace('\'', '')
conts = conts.replace('!', '')
words = conts.split(' ')
words = [word for word in words if len(word) >= 1]

ngrams = []
for i in range(len(words)-1):
    ngrams.append([words[i], words[i+1]])
# print(ngrams)
ngrams_list = []
for ngram in ngrams:
    ngrams_list.append(ngram[0] + ' ' + ngram[1])
count_ngram = Counter(ngrams_list)
print(count_ngram)

file.close()
```
