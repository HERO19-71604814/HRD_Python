# <농산물 가격 데이터 가져오기>
## - https://www.kamis.or.kr/customer/price/agricultureRetail/period.do
## - 2013년 ~ 2022년 대전 배추 가격 엑셀데이터 다운로드 후 수정

```python
import csv

prod = [] # 데이터 담을 공간 생성
for i in range(10):
    file = open('농산물/가격정보' + str(i + 2013) + '.csv') # range(10) -> 1~10 -> i -> str(i + 2013) -> 2013,2014,...2022 
    data = csv.reader(file) # 파일 읽어 오기
    
    next(data) # 구분, 평균, 평년이 적힌 1행 제외
    for item in data:
        # 평균
        if item[1].strip().isdigit():
            avg = int(item[1])
        else:
            avg = 0
        # 평년
        if item[2].strip().isdigit():
            avg_year = int(item[2])
        else:
            avg_year = 0
            
        prod.append([str(i + 2013) + '-' + str(item[0][:2]) + '-' + str(item[0][4:6]), avg, avg_year])
    file.close()
```

#### - isdigit()함수
문자열이 '숫자'로만 이루어져있는지 확인하는 함수 입니다.

문자가 '단 하나'라도 있다면 False를 반환하고,
모든 문자가 '숫자'로만 이루어져있으면 True를 반환합니다.

#### - strip()함수
문자열내에서 원하는 문자열 또는 공백을 모두 제거한다.

```python
for row in prod:
    print(row)
```

# - 데이터 베이스에 데이터 삽입
daejeon_temp 테이블에 number타입의 prod_avg와 prod_avg_year를 추가 생성 이후 데이터 삽입

```python
import cx_Oracle

dsn = cx_Oracle.makedsn('localhost', 1521, 'xe')
db = cx_Oracle.connect('PYTHON', '1234', dsn)
cursor = db.cursor()

for row in prod:
    sql = "update daejeon_temp set prod_avg = :1, prod_avg_year = :2 " 
    sql += "where to_char(wdate, 'YYYY-MM-DD') = :3"
#     print(sql)
    data_list = [row[1], row[2], row[0]]
    print(data_list)
    cursor.execute(sql, data_list)           
    cursor.execute('commit')
    
```
