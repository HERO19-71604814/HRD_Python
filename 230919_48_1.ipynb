## - 전체데이터 가져오기

```python
import cx_Oracle

# DB 계정 접속
dsn = cx_Oracle.makedsn('localhost', '1521', 'xe')
conn = cx_Oracle.connect('PYTHON', '1234', dsn)

cursor = conn.cursor()

sql = "select * from daejeon_temp"
rs = cursor.execute(sql)

data_list = []

# data_list 데이터 삽입
for record in rs:
    day = str(record[0])[:10]
    avg_temp = record[1]
    high_temp = record[2]
    row_temp = record[3]
    prec = record[4]
    pred_avg = record[5]
    pred_avg_year = record[6]
    data_list.append([day, avg_temp, high_temp, row_temp, prec, pred_avg, pred_avg_year])

# 삽입된 데이터를 출력
for row in data_list:
    print(row)
```

## - None값 대신 전이나 후에 배추 값이 있는 데이터로 삽입

```python
if data_list[0][5] == None: # PROD_AVG의 값이 None일 경우
    for i in range(len(data_list)): # 전채 개수 만큼 반복
        if data_list[i][5] != None: # PROD_AVG의 값이 None이 아닐 경우
            data_list[0][5] = data_list[i][5] # PROD_AVG의 None값에 삽입 
            data_list[0][6] = data_list[i][6] # PROD_AVG_YEAR의 None값에 삽입
            break
            
for i in range(len(data_list)):# 전채 개수 만큼 반복
    if data_list[i][5] == None: # PROD_AVG의 값이 None일 경우
        data_list[i][5] = data_list[i-1][5] 
        data_list[i][6] = data_list[i-1][6]
```

## - 최근 10년 년도별 4,5,6,7,8월별 기온과 강수량,배추값 평균

```python
month_result = {}
for row in data_list:
    if row[0][5:7] == '04' or row[0][5:7] == '05' or row[0][5:7] == '06' or row[0][5:7] == '07' or row[0][5:7] == '08': 
        if row[0][0:7] not in month_result:
            month_result[row[0][0:7]] = []
        month_result[row[0][0:7]].append(row[1:])
        
#     <원래 코드>
#     if row[0][:4] == '2022':      
#         if row[0][5:7] not in month_result:
#             month_result[row[0][5:7]] = []
#         month_result[row[0][5:7]].append(row[1:])

# print(month_result)

# 평균 구하기
month_avg = {}
for key, value in month_result.items():
    if key not in month_avg:
        month_avg[key] = [0, 0, 0] # 초기 값
    for row in value:
        month_avg[key][0] += row[0]
        month_avg[key][1] += row[3]
        month_avg[key][2] += row[4]
# print(month_avg)
        
# 소수점 줄이기
for key, value in month_avg.items():
    for i in range(len(value)):        
         month_avg[key][i] = round(month_avg[key][i] / len(month_result[key]), 2)

# 최종 결과 출력
for key, value in month_avg.items():        
    print(key, value)
```
