```python
import pytesseract
from PIL import Image

# tesseract가 설치된 파일에서 응용프로그램 불러오기
pytesseract.pytesseract.tesseract_cmd = r'D:\Program Files\Tesseract-OCR\tesseract.exe'

# 한국어 포함된 이미지를 텍스트화
im = Image.open('imgs/test.jpg')
img_text = pytesseract.image_to_string(im, lang='kor')
print("<한국어가 포함된 이미지 텍스트화>\n")
print(img_text)
print()

# 영어 이미지를 텍스트화
im1 = Image.open('imgs/testen.png')
img_text1 = pytesseract.image_to_string(im1, lang='eng')
print('<영어 이미지를 텍스트화>\n')
print(img_text1)

```
