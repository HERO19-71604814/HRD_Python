```python
import seaborn as sns
data = sns.load_dataset('penguins')
data
```

```python
col_to_stay = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']
cols = col_to_stay + [col for col in data if col not in col_to_stay]
data = data[cols]
data
```
