### - 팽귄 데이터 가져오기

```python
import seaborn as sns

df = sns.load_dataset('penguins')
df.loc[[1,3], ['bill_length_mm']] = 50
df.loc[2, ['bill_length_mm']] = 60
df.loc[:4, 'bill_depth_mm'] = 60
df.loc[df['bill_length_mm'] >= 50] = 100
df
```
