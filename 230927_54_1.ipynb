```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic
```

## ■ 단일 조건식을 활용하여 데이터 필터링하기(159p)

```python
titanic[titanic['pclass'] == 3]
```

## ■ 다중 조건식을 활용하여 데이터 필터링하기(159p)

```python
titanic[(titanic['pclass'] == 3) & (titanic.sex == 'male')]
```

## 3.1.2 loc 조건부 필터링(160p)

```python
titanic.loc[titanic.fare > 300]
```

```python
titanic.loc[titanic.fare > 300, :]
```

```python
titanic.loc[titanic.fare > 300, ['age', 'fare', 'class']]
```

## 3.1.3 특정 값 포함 여부 필터링(161p)

```python
titanic[titanic.embarked.isin(['S', 'C'])]
```

### ■ 다중 컬럼에서 특정 값이 포함된 데이터 필터링하기(162p)

```python
titanic[(titanic.pclass.isin([1,3])) & (titanic.embarked.isin(['S', 'C']))]
```

```python
filter_pclass = titanic.pclass.isin([1,3])
filter_embarked = titanic.embarked.isin(['S', 'C'])
titanic[filter_pclass & filter_embarked]
```

## 3.1.4 쿼리를 사용하여 데이터 필터링하기(162p)

```python
titanic.query("pclass == [1,3] & embarked == ['S', 'C']")
```

## 3.1.5 결측값을 제외하고 데이터 필터링하기(163p)

```python
# titanic[titanic.age.notnull()]
titanic[titanic.age.isnull()]
```

## 3.1.6 특정 문자가 포함된 데이터 필터링하기(164p)

```python
titanic[(titanic.embark_town.str.contains('South')) & titanic.embark_town.notnull()]
```

## 3.1.7 다양한 기준으로 데이터 정렬하기(165p)

```python
titanic.sort_values(by=['fare'], ascending = False)
```

```python
len(titanic[titanic.fare == 0])
titanic[titanic.fare == 0]
```

```python
titanic.sort_values(by=['fare', 'age'], ascending = False)
```

```python
titanic[titanic['sex'] == 'female'].nlargest(n=10, columns='age')
```

```python
titanic[titanic['sex'] == 'female'].nlargest(n=10, columns='age').sort_index()
```
