### 3.3.2 이상값 시각화하기(181p)

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 4))
sns.boxplot(x=titanic.fare)
```

```python
sns.boxplot(x=titanic.age)
```

### 3.3.3 IQR 기법으로 이상값 확인하기(183p)

```python
import numpy as np

q1 = np.quantile(titanic.fare, 0.25)
q3 = np.quantile(titanic.fare, 0.75)
iqr = q3 - q1
print('q1 : {} \nq3 : {} \niqr : {}\n'.format(q1, q3, iqr))

lower = q1 - iqr * 1.5
upper = q3 + iqr * 1.5
print('하위이상값 : {} 미만'.format(lower))
print('상위이상값 : {} 이상'.format(upper))
```
