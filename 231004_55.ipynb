### 3.4.2 문잣값 교체하기 (191p)

```python
titanic['title'] = titanic.Name.str.split().str[1]
titanic.title.value_counts()

rare_name = list(titanic.title.unique())

del rare_name[0:4]
rare_name
# rare_name.remove('Master.')

titanic['title'] = titanic['title'].replace(rare_name, 'Rare', regex=False)
titanic['title']
```

```python
titanic.title = titanic.title.str.replace('.', '', regex=False)
# titanic.title.value_counts()
titanic.title
```

```python
titanic.title.value_counts()
```

### 3.4.3 정규 표현식 가이드(194p)

### ■ 판다스 정규 표현식 기초 활용 예제 (197p)

```python
titanic.title = titanic.Name.str.extract(r'([A-Za-z]+)\.', expand=False)
titanic.title.value_counts()
```

```python
# 이름이 알파벳이 Z로 시작하는 승객의 데이터 추출
titanic[titanic.Name.str.count(r'(^Z.*)') == 1]
```

```python
# 괄호가 포함되어져 있는 데이터 추출
titanic[titanic.Name.str.contains(r'\(') == True]
```

## 3.5 카테고리 데이터 처리(200p)

### 3.5.1 숫자 타입 데이터를 카테고리 타입으로 만들기
### ■ 구간형으로 범주화 하기 (201p)

```python
# 최댓값과 최솟값
print("최솟값 : {}".format(titanic.Age.min()))
print("최댓값 : {}".format(titanic.Age.max()))

# 아기의 생존 여부
titanic[titanic.Age == 0.42]

# 80세 생존 여부
# titanic[titanic.Age == 80]
```

```python
bins = [0, 9, 18, 40, 60, 81] # 나이를 구분하는 포인트 지정
labels = ['어린이', '청소년', '20/30대', '40/50대', '60대이상'] # bins보다 1개 더 적게 정의
titanic['Age_band_cut'] = pd.cut(titanic['Age'], bins = bins, labels = labels)
titanic.Age_band_cut.value_counts()
```

```python
# 판다스의 style 모듈과 background_gradient() 함수
titanic.Age_band_cut.value_counts().to_frame().style.background_gradient(cmap='pink_r')
```

### ■비율형으로 범주화 하기(203p)

```python
# qcut() 함수
# 비율기준으로 구간을 5분할 설정 
titanic['Age_band_pct'] = pd.qcut(titanic['Age'], q=5, labels=labels)
titanic['Age_band_pct'].value_counts()
```

### 3.5.2 카테고리 데이터에 순서 만들기(205p)

```python
titanic['Pclass'] = titanic.Pclass.astype('category')
titanic['Pclass'] = titanic.Pclass.cat.set_categories([3, 2, 1], ordered=True)

# sort_values()함수
titanic.Pclass.sort_values()
```
