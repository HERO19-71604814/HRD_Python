## 4.1 데이터 병합(214p)

### 4.1.1 데이터를 위아래로 연결하기(214p)

```python
import pandas as pd

menu_1 = pd.Series(['파스타', '라면', '냉면'], index = [1, 2, 3])
menu_2 = pd.Series(['돈까스', '피자', '치킨'], index = [4, 5, 6])

# concat() 함수: 어떤축기준으로 병합
pd.concat([menu_1, menu_2])
```

```python
data_1 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '카테고리1' : ['일식', '양식', '일식', '양식', '중식']
})

data_2 = pd.DataFrame({
    '음식명' : ['갈비탕', '냉면', '짜장면', '파스타', '라멘'],
    '카테고리2' : ['한식', '한식', '중식', '양식', '일식']
})

# ignore_index=True
pd.concat([data_1, data_2], ignore_index=True)
```

### 4.1.2 컬럼 기준으로 데이터 병합(좌우로 연결)하기 (220p)

### ■ 데이터를 좌우로 연결하기

```python
data_1 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '카테고리' : ['일식', '양식', '일식', '양식', '중식']
})

data_2 = pd.DataFrame({
    '음식명' : ['갈비탕', '냉면', '짜장면', '파스타', '라멘'],
    '카테고리' : ['한식', '한식', '중식', '양식', '일식']
})

# axis = 1
pd.concat([data_1, data_2], axis = 1)
```

### ■특정 컬럼을 기준으로 데이터 병합하기(220p)

```python
data_1 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '카테고리' : ['일식', '양식', '일식', '양식', '중식']
})

data_2 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '지역명' : ['서울', '대전', '대구', '부산', '광주']
})

# merge()함수: 교집합 방식으로 작동
pd.merge(data_1, data_2)
```

```python
data_1 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '카테고리' : ['일식', '양식', '일식', '양식', '중식']
})

data_2 = pd.DataFrame({
    '음식명' : ['돈까스', '라면', '초밥', '김밥', '탕수육'],
    '지역명' : ['서울', '대전', '대구', '부산', '광주']
})

# how = 'outer': 합집합 방식을 사용
pd.merge(data_1, data_2, how = 'outer')
```

```python
data_1.merge(data_2, how = 'outer')
```

### ■ 특정 컬럼을 기준으러 데이터 병합(224p)

```python
data_1 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '카테고리' : ['일식', '양식', '일식', '양식', '중식']
})

data_2 = pd.DataFrame({
    '음식명' : ['돈까스', '라면', '초밥', '김밥', '탕수육'],
    '지역명' : ['서울', '대전', '대구', '부산', '광주']
})

# join() 함수
data_1.join(data_2, lsuffix='_left', rsuffix='_right')

# data_1.merge(data_2)
```

## 4.1.3 key 컬럼에 값이 같은 데이터가 여러 개 있는 경우 데이터 병합하기(225p)

```python
data_1 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '가격' : [10000, 15000, 12000, 20000, 25000]
})

data_2 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '위치' : ['서울', '대전', '대구', '부산', '광주']
})

data_1.merge(data_2)
```

## 4.1.4 컬럼명이 동일하지만 key 컬럼이 되면 안 되는 경우 데이터 병합하기(227p)

```python
data_1 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '가격' : [10000, 15000, 12000, 20000, 25000],
    '구분' : ['양식', '양식', '일식', '양식', '중식'] 
})

data_2 = pd.DataFrame({
    '음식명' : ['돈까스', '피자', '초밥', '치킨', '탕수육'],
    '위치' : ['서울', '대전', '대구', '부산', '광주'],
    '구분' : ['양식', '양식', '일식', '양식', '중식'] 
})

# data_1.merge(data_2, on = '음식명')
data_1.merge(data_2)
```
