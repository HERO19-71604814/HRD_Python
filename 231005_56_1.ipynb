```python
import pandas as pd

subway = pd.read_csv('data/승하차인원.csv', encoding='cp949')
subway
```

## - 역별 8시대 승하차 인원 집계(합계)

```python
# pivot_table()
import numpy as np
subway.pivot_table(index='역명', columns='구분', values='08-09시')
# 매개변수 aggfunc를 생략하면 평균값
```

```python
import numpy as np
subway.pivot_table(index='역명', columns='구분', values='08-09시', aggfunc='max')

# aggfunc : 
# 합계 : sum
# 평균 : mean
# 표준편차 : std
# 카운트 : count
# 최대값 : max
#최속값 : min
```

```python
import numpy as np
subway.pivot_table(index='역명', columns='구분', values='08-09시', aggfunc='mean')
```

```python
import numpy as np
subway_sum = subway.pivot_table(index='역명', columns='구분',
                   values=['07-08시','08-09시','18-19시','19-20시'], aggfunc='sum')
subway_sum[subway_sum['07-08시']['승차'] >= 50000]
```
