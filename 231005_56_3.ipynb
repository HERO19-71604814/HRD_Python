## 5.1 그룹 연산의 이해
## 5.1.1 groupby() 연산 기초(254p)

```python
# 다이아몬드 데이터 셋
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
# groupby() 함수 
grouped = diamonds['price'].groupby(diamonds['clarity'])
grouped.mean()
```

```python
grouped = diamonds['price'].groupby([diamonds['color'],diamonds['clarity']]).mean()
grouped
```

```python
diamonds.groupby(['cut']).mean()
```

```python
di = diamonds.groupby(['carat']).size()
di
```

```python
diamonds.groupby('cut')['price'].mean()
```

```python
diamonds.groupby('color')['price'].mean()
```

```python
diamonds.groupby(['cut','color'])['price'].mean().unstack()
```

```python
diamonds.groupby(['cut','color','clarity'])['price'].mean().unstack()
```
