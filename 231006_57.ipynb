```python
import pandas as pd 
import numpy as np

food = pd.DataFrame(np.random.randn(5, 5),
                    columns = ['피자', '햄버거', '돈까스', '스테이크', '초밥'],
                    index = ['지민', '뷔', '정국', '진', 'RM'] )

food.iloc[1:2, [2,4]] = np.nan
food
```

```python
maping = {'피자' : '양식',
          '햄버거' : '양식',
          '돈까스' : '일식',
          '초밥' : '일식',
          '스테이크' : '양식',
          '짜장면' : '중식'}
```

```python
food.groupby(maping, axis=1).sum()
```

```python
# 시리즈 타입으로 변경
maping_series = pd.Series(maping)
maping_series
```

```python
food.groupby(maping_series, axis=1).count()
```

```python
food.groupby(len).sum()
```

### ■ 컬럼에 다양한 함수를 적용하는 방법(262p)

```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips
```

```python
# 손님이 계산한 총 액수에서 팀의 비중 구하기
tips['tips_percent'] = tips['tip'] / tips['total_bill']
tips
```

```python
# 요일과 흡연여부 기준으로 데이터를 다시 그루핑
tips_byday = tips.groupby(['day', 'smoker'])
tips_byday['tips_percent']
```

```python
# age 함수
tips_byday_percent = tips_byday['tips_percent']
# tips_byday_percent.agg([('합계','sum'), ('평균', 'mean'), 'max', 'min'])

# tips_byday_percent.mean()

func_list = [('합계','sum'), ('평균', 'mean'), 'max', 'min']
tips_byday_percent.agg(func_list)
```

```python
tips_byday.agg({'tip' : 'max', 'size' : 'sum'})
```
