# 대전기온.csv 파일을 읽어서 월별 컬럼에 대한 평균값을 구하시오.

```python
import pandas as pd

data = pd.read_csv('data/대전기온.csv', encoding='cp949')
data
```

```python
data['일자'] = pd.to_datetime(data['일자'])
data['월'] = data['일자'].dt.month
data['요일'] = data['일자'].dt.weekday
weekday_list = ['월', '화', '수', '목', '금', '토', '일']
data['요일'] = data.apply(lambda x : weekday_list[x['요일']], axis=1)
data

# type(data['월'])
# type(data['월'].str)
```

```python
data_month = data.groupby(data['월'])
data_month.mean()
```

# 요일별 컬럼에 대한 평균값을 구하시오.

```python
data_month = data.groupby(data['요일'])
data_month.mean()
```
