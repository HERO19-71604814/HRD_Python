# 어쩌다 데이터 분석 with 파이썬
## 5.1.3 groupby-apply 작업의 개념
### ■ 사용자 정의 함수에 apply 활용 예(268p)

```python
import seaborn as sns

penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins
```

```python
def top(data, n=5, columns='body_mass_g'):
    return data.sort_values(by=columns)[-n:]
```

```python
# 성별 그룹화 => 상위 n개 출력
penguins.groupby('sex').apply(top, n=10, columns='bill_length_mm')
```

### ■describe() 함수에 apply 활용 예(271p)

```python
# 성별기준으로 상세 분포 확인하기
penguins.groupby('sex')['body_mass_g'].describe()
```

```python
# 행 변환 후 확인하기
penguins.groupby('sex')['body_mass_g'].describe().unstack()
```

```python
penguins.groupby('sex')['body_mass_g', 'bill_length_mm'].apply(lambda x: x.describe())
```

### ■ 버킷 분석에 apply 활용 예(273p)

```python
import pandas as pd

penguins['body_mass_g_bin'] = pd.cut(penguins['body_mass_g'], bins=5)
# penguins

penguins['body_mass_g_bin'].value_counts()
```

## 6.1.1 datetime 라이브러리 데이터 타입(284p)

```python
import datetime

# 현재 일자
today = datetime.date.today()
print("현재일자 : {}".format(today))

# 생년월일
birthday = datetime.date(1999, 2, 16)
print("생년월일 : {}".format(birthday))

# 태어난 날로 부터 오늘까지 살아온 일수
delta = today - birthday
print("태어난 일로부터 {}일 째".format(delta.days))
```

# - 문제 풀어 보기
1년간 데이터 측정값을 랜덤하게 구하고자 한다 <br>
측정항목은 'A', 'B', 'C' 이고 <br>
값의 범위는 <br>
'A': -1 ~ 1 <br>
'B': 10 ~ 100 <br>
'C': 1000 ~ 2000 <br>
측정 주기는 5시간 단위로 <br>

```python
import numpy as np

# 2023년 전체 기간
start = datetime.datetime(2023, 1, 1)
end = datetime.datetime(2023, 12, 31)

# 5시간 단위로 측정
index = pd.date_range(start, end, freq='5H')
# print(len(index))
# print(index)

a = np.random.rand(len(index) * 2 -1)
b = np.random.randint(10, 20, size = len(index))
c = np.random.randint(1000, 2000, size = len(index))

print(a)
print(b)
print(c)
```
