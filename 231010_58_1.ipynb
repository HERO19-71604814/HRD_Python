# 어쩌다 데이터 분석 with 파이썬
### 6.3.2 시간값 추출하기(306p)

```python
import pandas as pd

df = pd.read_csv('data/AirQualityUCI.csv', sep = ';')

# data_air = [[날짜, 시간, 온도, 상대습도, 절대습도]]
data_air = df[['Date', 'Time', 'T', 'RH', 'AH']]
# data_air

# 모든 컬럼의 데이터 타입을 확인
# data_air.dtypes

# float 타입으로 변경
data_air['T'] = data_air['T'].str.replace(',', '.').astype(float)
data_air['RH'] = data_air['RH'].str.replace(',', '.').astype(float)
data_air['AH'] = data_air['AH'].str.replace(',', '.').astype(float)

# Date + Time 결합하기
data_air['Date_Time'] = data_air['Date'] + ' ' + data_air['Time']
data_air
```

```python
# na 값 제거
data_air = data_air.dropna()
# data_air['Date_Time'].dtype

# datatime 형식으로 변경
data_air['Date_Time'] = pd.to_datetime(data_air['Date_Time'], format='%d/%m/%Y %H.%M.%S')
data_air
```

### ■ 6.3.3 날짜/시간 속성 활용 예제 (311p)

```python
# na 값 제거
data_air = data_air.dropna()
# data_air['Date_Time'].dtype

# datatime 형식으로 변경
data_air['Date_Time'] = pd.to_datetime(data_air['Date_Time'], format='%d/%m/%Y %H.%M.%S')

data_air['year'] = data_air['Date_Time'].dt.year
data_air['month'] = data_air['Date_Time'].dt.month
data_air['day'] = data_air['Date_Time'].dt.day
data_air['weekday'] = data_air['Date_Time'].dt.weekday
data_air['hour'] = data_air['Date_Time'].dt.hour

data_air
```

```python
data_air['hour_4cut'] = pd.cut(data_air['hour'], 4, labels=['night', 'morning', 'afternoon', 'evening'])

data_air
```

### ■ 요일과 시간대를 기준으로 평균과 표준편차 계산하기(312p)

```python
data_byhour_4cut = data_air.groupby(['weekday', 'hour_4cut'])
data_byhour_4cut.agg(['mean', 'std'])[['T', 'RH', 'AH']]
```

### ■ 월과 요일을 기준으로 최댓값 구하기(313p)

```python
data_air_bymonth = data_air.groupby(['month', 'weekday'])
# data_air_bymonth.agg('max',['T', 'RH', 'AH'])
data_air_bymonth.agg({'T' : max, 'RH' : max, 'AH' : max})
```
