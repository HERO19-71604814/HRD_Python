## - 대전기온 데이터 가져오기

```python
import pandas as pd
import datetime

data = pd.read_csv('data/대전기온.csv', encoding='cp949')
data
```

## - 월, 주별 최고기온의 평균과 최대값을 구하기오.

```python
# 일자 타입을 datetime으로 변경
data['일자'] = pd.to_datetime(data['일자'], format='%Y-%m-%d')

# 월, 주 컬럼 추가
data['월'] = data['일자'].dt.month
data['주'] = data['일자'].dt.weekday

data_bymax = data.groupby(['월', '주'])
# 생략되는 부분 확인하기
pd.set_option('display.max.rows', 300)
data_bymax.agg(['mean', 'max'])['최고기온'].head(84)
```
