```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_location = 'C:/windows/fonts/H2GPRM.TTF'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = font_name)

plt.title('제목', color = 'blue', size = 30)
# plt.plot(df.kor, label='kor', c='r', ls='solid')
# plt.plot(df.eng, label='eng', c='b', ls='dashed')
# plt.plot(df.mat, label='mat', c='g', ls='dotted')

plt.plot(df.kor, label='kor', c='r', ls='-', marker='o')
plt.plot(df.eng, label='eng', c='b', ls='--', marker='^')

# plt.plot(df.mat, label='mat', c='g', ls=':', marker='s')
plt.plot(df.mat, 'gs:', label='mat')

# plt.ylim([0,100])
plt.axis([0, 6, 0, 100])
plt.legend(loc=(1.0, 1.0))
plt.show()
```

### ■ 여러 그래프 동시에 그리기(326p)

```python
plt.figure(figsize=(20, 10))
plt.subplots_adjust(hspace=1, wspace=0.3)

plt.subplot(2, 2, 1)
plt.title('국어')
plt.plot(df.kor)

plt.subplot(2, 2, 2)
plt.title('영어')
plt.plot(df.eng)

plt.subplot(2, 2, 3)
plt.title('수학')
plt.plot(df.mat)

plt.show()
```
