# - 대전기온 데이터와 글꼴 데이터 가져오기 

```python
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_location = 'C:/windows/fonts/H2GPRM.TTF'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = font_name)

data = pd.read_csv('data/대전기온.csv', encoding='cp949')
```

# - 대전 연도, 월별 평균기온 그래프

```python
data['일자'] = pd.to_datetime(data['일자'], format='%Y-%m-%d')

data['년'] = data['일자'].dt.year
data['월'] = data['일자'].dt.month
data

data_mean_month = data.groupby(['년', '월'])
mean_data = list(data_mean_month.mean()['평균기온'])

# data_mean_month.agg('mean')[['년', '월']]
plt.plot(mean_data)
plt.show()
```
