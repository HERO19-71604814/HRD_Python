# 어쩌다 데이터 분석 with 파이썬
### 7.1.2 데이터 분포 확인 : 히스토그램(329p)

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_location = 'c:/windows/fonts/H2GTRE.TTF'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = font_name)

iris = sns.load_dataset('iris')
iris
```

```python
plt.figure(figsize=(15,8))
plt.title('IRIS')
plt.hist(iris.sepal_length, color='b', rwidth=0.8, histtype='bar')
plt.hist(iris.petal_length, color='r', rwidth=0.5, histtype='bar')
# plt.hist(iris.sepal_width, color='r', rwidth=0.8, histtype='bar')
# plt.plot(iris.petal_length, color='r')
plt.show()
```

### ■ 누적 히스토그램 시각화(330p)

```python
plt.hist(iris.sepal_length, cumulative=True, label='cumulative', color="g")
plt.hist(iris.sepal_length, cumulative=False, label='not cumulative', color="b")
plt.legend(loc='upper left')
plt.show()
```

### - 히스토그램 4가지 타입 선택(331p)

```python
fig = plt.figure(figsize=(20, 15))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.title.set_text('sepal_length')
ax1.hist(iris.sepal_length, color='r', histtype='bar', rwidth=0.8)
ax2.title.set_text('sepal_width')
ax2.hist(iris.sepal_width, color='b', histtype='barstacked', rwidth=0.8)
ax3.title.set_text('petal_length')
ax3.hist(iris.petal_length, color='g', histtype='step', rwidth=0.8)
ax4.title.set_text('petal_width')
ax4.hist(iris.petal_width, color='y', histtype='stepfilled', rwidth=0.8)
plt.tight_layout()
plt.show()

```
