## - 대전 서구 동별 연령대별 인구수 그래프 만들어보기

```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_location = 'c:/windows/fonts/H2GTRE.TTF'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = font_name)

data = pd.read_excel('data/2023연령별인구.xlsx')
deajeon_data = data[data['행정기관'].str.contains('대전광역시')]
deajeon_data
```

```python
# 서구에 대한 전체 데이터 
deajeon_data_sgu = deajeon_data[
    deajeon_data['행정기관'].str.split().str[1] == '서구' 
]
deajeon_data_sgu
```

```python
gu_ST = deajeon_data_sgu.T # .T함수로 행과 열을 교체
gu_ST = gu_ST.rename(columns=gu_ST.iloc[0].str.split().str[2]) # 컬럼들의 이름을 각각의 구로 변경
# 필요없는 부분을 제거
gu_ST = gu_ST.drop(gu_ST.index[0:3]) # 행정기관, 총 인구수, 연령구간인구수 제거
gu_ST = gu_ST.drop(gu_ST.index[11:]) # 남,여 인구수 + 연령인구수 제거
gu_ST
```
