# - 2023년 대전의 연령별 인구 데이터 가져오기

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd

font_location = 'c:/windows/fonts/H2GTRE.TTF'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = font_name)

data = pd.read_excel('data/2023연령별인구.xlsx')
deajeon_data = data[data['행정기관'].str.contains('대전광역시')]
deajeon_data
```

### - 각각의 구의 동들에 대한 데이터 가져오기

```python
deajeon_data_dong= deajeon_data[
    deajeon_data['행정기관'].str.split().str[2].notnull() 
]
deajeon_data_dong
```

### - 막대 그래프로 출력
- xticks(rotation=00) 그래프에 밑에 이름을 45도 기울여서 출력

```python
fig = plt.figure(figsize=(50,15))
plt.bar(deajeon_data_dong['행정기관'].str.split().str[2], 
        deajeon_data_dong['총 인구수'])
plt.xticks(deajeon_data_dong['행정기관'].str.split().str[2],rotation=45)
plt.show()
```

### - 구와 동으로 나누어진 컬럼을 추가

```python
deajeon_data_dong['구'] = deajeon_data_dong['행정기관'].str.split().str[1]
deajeon_data_dong['동'] = deajeon_data_dong['행정기관'].str.split().str[2]
# print(deajeon_data_dong['구'])
# print(deajeon_data_dong['동'])
```

### 행정기관[0], 총 인구수[1], 연령구간인구수[2] 제거<br> 이 후에 남녀 인구수[11] ~ 제거 

```python
deajeon_data_dong = deajeon_data_dong.drop(deajeon_data_dong.columns[0:3], axis=1)
deajeon_data_dong = deajeon_data_dong.drop(deajeon_data_dong.columns[11:37], axis=1)
deajeon_data_dong
```

```python
# deajeon_data_dong의 전체컬럼
deajeon_data_dong.columns
```

### - 구, 동을 멀티 인덱스 지정

```python
deajeon_gu_dong_idx = deajeon_data_dong.set_index(keys=['구', '동'])
deajeon_gu_dong_idx
```

### - 대전의 인구수에 대한 데이터 그래프
- .xs(값) 함수는 멀티인덱스 객체에 대해서 하위 분류를 출력하는 메서드입니다.
- .T 함수는 (전치)행과 열 바꾸기

```python
gu_list = ['동구', '중구', '서구', '유성구', '대덕구']
font_dict = {'fontsize' : 24, 'fontweight' : 'bold'}

# 대전의 구에 대한 반복
for i in range(len(gu_list)):
    gu = deajeon_gu_dong_idx.xs(gu_list[i])
    gu_T = gu.T
    dong_list = gu_T.columns
    print()
    print('\033[1m'  + '<대전광역시 ' + gu_list[i] + '>' + '\033[0m')
    print()
    
    # 대전의 구의 동에 대한 반복
    for j in range(len(dong_list)):
        plt.figure(figsize=(20, 8))
        plt.title(dong_list[j], font_dict)
        plt.ylim([0, 10000])
        plt.bar(gu_T.index, gu_T[dong_list[j]])
        plt.show()
        
```
