### 7.1.4 수치형 데이터 상관관계: 산점도(334p)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import matplotlib.font_manager as fm
import pandas as pd

font_location = 'c:/windows/fonts/H2GTRE.TTF'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = font_name)

iris = sns.load_dataset('iris')
iris.head()
```

```python
import numpy as np
area = (18 * np.random.rand(len(iris.sepal_length))) **2
colors = np.random.rand(len(iris.sepal_length))
plt.scatter(x=iris.sepal_length, y=iris.petal_length, s=area, c=colors, 
            alpha=0.75, cmap='Set1_r')
plt.show()
```

```python
# %matplotlib tk 외부에서 그릴 때(새로운 창)

plt.scatter(x=iris.sepal_length, y=iris.petal_length, s=area, c='hotpink', 
            alpha=0.75)
plt.scatter(x=iris.sepal_width, y=iris.petal_width, s=area, c='blue', 
            alpha=0.75)
plt.scatter(x=iris.sepal_length, y=iris.petal_width, s=area, c='orange', 
            alpha=0.75)
plt.show()
```

```python
%matplotlib inline # 내부에서 그릴 때

plt.scatter(x=iris.sepal_length, y=iris.petal_length, s=area, c='hotpink', 
            alpha=0.75)
plt.scatter(x=iris.sepal_width, y=iris.petal_width, s=area, c='blue', 
            alpha=0.75)
plt.show()
```

### 7.2 시본 데이터 시각화
### 7.2.1 수치형 데이터: 라인그래프(342p)

```python
sns.set_style('whitegrid') # 그래프 스타일 테마 지정
```

```python
# 비행기 데이터 셋
flights = sns.load_dataset('flights')
flights.head()
```

```python
# x = 년도, y = 승객수
sns.lineplot(x='year', y='passengers', data=flights)
```

- query() 함수를 사용하여 9월 데이터만 한정해서 시각화

```python
september_flights = flights.query("month == 'Sep'")
sns.lineplot(x='year', y='passengers', data=september_flights)
```

- 매개변수 hue에 구분하고자 하는 변수 입력 -> 전체 평균값을 1개로 표현한 라인을 월 기준으로 분리

```python
sns.lineplot(x='year', y='passengers', hue='month', data=flights)
```

#### tips 데이터를 활용한 relplot() 함수

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='day', y='tip', kind='line', col='sex', hue='smoker', ci=None, data=tips)
```

### 7.2.2 데이터 분포 확인: 히스토그램(345p)
- 팽귄 데이터를 활용한 히스토그램

```python
penguins = sns.load_dataset('penguins')

sns.histplot(x='flipper_length_mm', data=penguins)
```

- bins는 막대그래프 수 증가 및 감소 시킬수 있다

```python
sns.histplot(x='flipper_length_mm', bins=20, data=penguins)
```

- binwidth는 너비를 조정할 수 있다.

```python
sns.histplot(x='flipper_length_mm', binwidth=2, data=penguins)
```

- 수평막대그래프로 그리기

```python
sns.histplot(y='flipper_length_mm', data=penguins)
```

- hue로 그룹별 세분화후 시각화

```python
sns.histplot(x='flipper_length_mm', hue='species', data=penguins)
```

- multiple='stack'을 사용하면 막대그래프가 차곡히 쌓인형태로 볼수 있다

```python
sns.histplot(x='flipper_length_mm', hue='species', multiple='stack', data=penguins)
```

- 히스토그램 여러 개 출력은 displot() 함수를 사용, col='분리기준 값'이 분리 기준이 된다

```python
sns.displot(x='body_mass_g', hue='species', col='sex', kind='hist', multiple='stack', data=penguins)
```

### 7.2.3 수치형 데이터 상관관계: 산점도(348p)
- tips 데이터셋을 사용

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

- hue로 점심과 저녁 시간대를 분리 후 출력

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

- 산점도의 사이즈 변경 -> size='적용할 변수', sizes=(도형의 최소, 최대 사이즈)

```python
sns.scatterplot(x='total_bill', y='tip', hue='size', 
                size='size', sizes=(20, 200), legend='full' , data=tips)
```

- 산점도여러 개 동시 출력 시 relplot함수 활용, 사용법은 displot()함수와 유사

```python
sns.relplot(x='total_bill', y='tip', col='time', hue='day', style='day',
            kind='scatter', data=tips)
```

### 7.2.4 범주형 데이터 개수 확인: 막대그래프 - countplot() (351p)

```python
sns.countplot(x='day', data=tips)
```

```python
sns.countplot(x='day', hue='time', data=tips)
```

- 한번에 여러 개의 countplot()을 시각화 할 때 catplot()함수를 사용

```python
sns.catplot(x='sex', hue='time', col='day', 
            data=tips, kind='count', height=5, aspect=.6)
```

### 7.2.5 범주형과 수치형 데이터: 막대그래프 - barplot() (352p)

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips, ci=65, palette='Blues_d')
```

```python
sns.catplot(x='day', y='total_bill', hue='smoker', col='time', 
            data=tips, kind='bar', palette='Blues_d', height=5, aspect=.9)
```
