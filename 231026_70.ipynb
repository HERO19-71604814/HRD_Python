## - 네트워크 그래프

```python
# 모듈 추가
import networkx as nx
import matplotlib.pyplot as plt
# 그래프 글자 출력 오류 해결 
import matplotlib
import matplotlib.font_manager as fm

# font_location = 'c:/windows/fonts/H2GTRE.TTF'
font_location = 'C:/Users/LGHERO/AppData/Local/Microsoft/Windows/Fonts/NanumGothic.TTF'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = font_name)
```

```python
# 그래프 생성
G = nx.Graph()
```

```python
# node 추가
G.add_nodes_from(['가', '나', '다', '라', '마', '바', '사', '아', '자', '차'])

# edge 추가
G.add_edges_from([('가', '차'), ('나', '라'), ('다', '라'), 
                  ('라', '바'), ('마', '바'), ('바', '자'), 
                  ('사', '아'), ('아', '가'), ('자', '나'), 
                  ('차', '다'), ('라', '아'), ('다', '마')])

# 네트워크 시각화
nx.draw_networkx(G, font_family = font_name, font_size=10)
plt.axis('off')
plt.show()
```

```python
# node, edge에 속성 부여하기
G.nodes['가']['gender'] = 'male'
G.nodes['나']['gender'] = 'female'
G.nodes['다']['gender'] = 'male'
G.nodes['라']['gender'] = 'male'
G.nodes['마']['gender'] = 'female'
G.nodes['바']['gender'] = 'male'
G.nodes['사']['gender'] = 'male'
G.nodes['아']['gender'] = 'female'
G.nodes['자']['gender'] = 'female'
G.nodes['차']['gender'] = 'male'


# [(1, 2), (1, 4), (2, 6), (3, 5), (5, 2), (4, 2)]
G['가']['차']['weight'] = '가, 차'
G['나']['라']['weight'] = '나, 라'
G['다']['라']['weight'] = '다, 라'
G['라']['바']['weight'] = ''
G['마']['바']['weight'] = ''
G['바']['자']['weight'] = ''
G['사']['아']['weight'] = ''
G['아']['가']['weight'] = ''
G['자']['나']['weight'] = ''
G['차']['다']['weight'] = ''
G['라']['아']['weight'] = '라, 아'
G['다']['마']['weight'] = '다, 마'
```

```python
print(nx.get_edge_attributes(G, 'weight'))
```

```python
G.nodes(data=True)
G.edges(data=True)
pos = nx.spring_layout(G)
# shell, spectral, circular, kamada_kawai, random
# pos = nx.shell_layout(G)
# pos = nx.spectral_layout(G)
# pos = nx.circular_layout(G)
# pos = nx.kamada_kawai_layout(G)
# pos = nx.random_layout(G)

nx.draw_networkx(G, pos, font_family = font_name, font_size=10) 
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.axis('off')
plt.show()
```

```python
# 그래프를 파일로 보내기
# nx.write_graphml(G, 'graph_test.graphml') # 읽으수있게 변환이 필요
```

```python
# node 속성에 따른 색표시
color_map = []
for n, d in G.nodes(data=True):
    if d['gender'] == 'female':
        color_map.append('pink')
    else:
        color_map.append('skyblue')

pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, node_color=color_map, font_family = font_name, font_size=10)

labels = nx.get_edge_attributes(G, 'weight')
plt.axis('off')
plt.show()
```
